```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('~/Downloads/dj_results.csv', index_col=0)
```

```python
df["date"] = pd.to_datetime(df["date"]).dt.date
```

```python
today = datetime.date.today()

this_month = today.month
this_year = today.year
currency = 7.78
```

```python
df.shape # how many rows
```

```python
df.head()
```

```python
df = df[
    (df["guarantee_year"]==this_year) & 
    (df["guarantee_month"]>= this_month - 3 ) &
    (df["date"] >= today - datetime.timedelta(days=7))
 ]  # i
```

```python
df.shape
```

```python
def f(df):
    return pd.Series({
        "min_hkd": df["lowest_price_hkd"].min(),
        "max_hkd": df["lowest_price_hkd"].max(),
        "min_hkd_today": df[df["date"]==today]["lowest_price_hkd"].min(),
        "count_t7": len(df),
        "count_today": len(df[df["date"]==today]),
        "std": df["lowest_price_hkd"].std()
    })
```

```python
# I care about for each ref
# what's the min/max they are asking, and what's the variance
res = df.groupby("ref").apply(f).reset_index()
```

```python
res.head()
```

```python
msrp = pd.read_csv("~/Downloads/Rolex Datejust Database - Sheet1.csv", skiprows=4)
```

```python
msrp['Reference'] = msrp['Reference'].apply(lambda x: x[1:])
```

```python
msrp['MSRP'] = msrp['MSRP'].apply(lambda x: float(x.strip('$').replace(",","")))
msrp['Tax-in Price'] = msrp['Tax-in Price'].apply(lambda x: float(x.strip('$').replace(",","")))
```

```python
combined = msrp[["Reference", "MSRP", "Tax-in Price"]].merge(res, left_on=['Reference'], right_on=['ref'])
```

```python
combined['min_usd'] = combined['min_hkd'] / currency
```

```python
combined['pre_tax_premium'] = combined['min_usd'] - combined['MSRP']
combined['post_tax_premium'] = combined['min_usd'] - combined['Tax-in Price']
```

```python
combined.sort_values('post_tax_premium', ascending=0)
```

```python
combined.to_csv("~/Downloads/combined.csv")
```
